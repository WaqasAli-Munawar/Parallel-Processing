Our goal is to use parallel processing to analyze how often data engineering job descriptions mention each skill in a list of skills, such as SQL, Postgres, Python, and others.

**Task**



**Answer**

```python
import pandas as pd
job_postings = pd.read_csv("DataEngineer.csv")
num_rows = job_postings.shape[0]
num_cols = job_postings.shape[1]
num_rows
```

Above we saw that the dataset contains a total of 2,528 data engineering job postings. Our goal for this file is to create a frequency table that counts mentions of particular skills in the job posting descriptions.

**Task**



**Answer**

```python
job_postings["Job Description"] = job_postings["Job Description"].str.lower()

frequency = {}
frequency["postgres"] = job_postings["Job Description"].str.count("postgres").sum()
frequency["sql"] = job_postings["Job Description"].str.count("sql").sum()

print(frequency)
```

Above, we started building a dictionary with the frequencies of different skills in data engineering job postings. Now we will complete that dictionary by calculating the frequency of all of the skills stored in the `Skills.csv` file. All skill names in the file are already lowercase.

We want to build a dictionary where the keys are the skill names and the values are the number of occurrences of each skill in the `Job Description` column of the `job_postings` DataFrame.

Later in the file, we'll learn how to use parallel processing to build the same dictionary. We'll measure the execution time of both solutions to see how parallel processing can help us calculate things quickly. For now, we'll write a solution that doesn't use parallel processing.

**Task**



**Answer**

```python
skills = pd.read_csv("Skills.csv")

frequency = {}
for skill_name in skills["Name"]:
    frequency[skill_name] = job_postings["Job Description"].str.count(skill_name).sum()
    
print(frequency["programming"])
```

Above we built a frequency table that counts every mention of each skill in the job descriptions. Our solution doesn't use parallel processing. The first step toward using parallel processing is to transform our code into a function.

These have the same name as the two pieces of data that we used above. However, they are not necessarily the same. The value of these arguments can be a subset of the data. When we use parallel processing, we'll divide the data into chunks and pass each chuck to the `count_skills` function.

**Task**



**Answer**

```python
import time

def count_skills(job_postings, skills):
    frequency = {}
    for skill_name in skills["Name"]:
        frequency[skill_name] = job_postings["Job Description"].str.count(skill_name).sum()
    return frequency


start = time.time()
count_skills(job_postings, skills)
end = time.time()

runtime = end - start
```

In the previous file, we learned how to read a DataFrame in chunks. This is useful when the dataset doesn't fit into memory. Now we'll learn how to divide a DataFrame that is already in memory into chunks.



```python
from IPython.display import Image

Image("5.1-m566.gif")
```

**Task**



**Answer**

```python
import math

def make_chunks(df, num_chunks):
    num_rows = df.shape[0]
    chunk_size = math.ceil(num_rows / num_chunks)
    return [df[i:i+chunk_size] for i in range(0, num_rows, chunk_size)]

skill_chunks = make_chunks(skills, 8)
```

We're now ready to apply parallel processing. We would do the following:

```python
import concurrent.futures

def increment(value):
    return value + 1

values = [1, 2, 3, 4, 5, 6, 7, 8]


with concurrent.futures.ProcessPoolExecutor() as executor:
    futures = [executor.submit(increment, value) for value in values]

results = [future.result() for future in futures]
```

We've managed to calculate the frequency table for the skills in the data engineering job postings using parallel processing. The last question that remains to answer is how much faster was it?

To help us measure the execution time gain, we've created a function `count_skills_parallel()`. Given the `job_postings` and the `skills` DataFrames and a number of processes `num_processes`, this function replicates what we've done in this file:

**Task**



**Answer**

```python
def count_skills(job_postings, skills):
    frequency = {}
    for skill_name in skills["Name"]:
        frequency[skill_name] = job_postings["Job Description"].str.count(skill_name).sum()
    return frequency

def count_skills_parallel(job_postings, skills, num_processes=4):
    # Calculate results using paralleld processing
    skill_chunks = make_chunks(skills, num_processes)
    with concurrent.futures.ProcessPoolExecutor() as executor:
        futures = [executor.submit(count_skills, job_postings, skill_chunk) for skill_chunk in skill_chunks]
    results = [future.result() for future in futures]
    # Merge results
    merged_results = {}
    for result in results:
        merged_results.update(result)
    return merged_results

import time

# Measure execution times here
start = time.time()
count_skills(job_postings, skills)
end = time.time()
time_normal = end - start

start = time.time()
count_skills_parallel(job_postings, skills)
end = time.time()
time_parallel = end - start

print(time_normal / time_parallel)
```

What an improvement! Above, we saw that the ratio was close to 2, which means that the parallel processing solution is about two times faster.
