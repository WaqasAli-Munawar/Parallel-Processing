```python
import math

def make_chunks(data, num_chunks):
    chunk_size = math.ceil(len(data) / num_chunks)
    return [data[i:i+chunk_size] for i in range(0, len(data), chunk_size)]

chunks = make_chunks([1, 2, 3, 4, 5, 6], 3)
```

Above we generalized the `make_chunks()` function so that it also works with other kinds of data, not only DataFrames (note that it also still works with DataFrames).

```python
import concurrent.futures

def map_parallel(mapper, data, num_processes):
    chunks = make_chunks(data, num_processes)
    with concurrent.futures.ProcessPoolExecutor() as executor:
        futures = [executor.submit(mapper, chunk) for chunk in chunks]
    return [future.result() for future in futures]

values = [1, 4, 5, 2, 7, 21,     \
          31, 41, 3, 40, 5, 14,  \
          9, 32, 12, 18, 1, 30,  \
          6, 19, 23, 35, 12, 13, \
          0, 12, 42, 41, 11, 9]

results = map_parallel(max, values, 5)
```

Now we're going to finalize our work and create a `map_reduce()` function that we can use to apply MapReduce to any dataset. Here's the information a user needs to provide:

**Answer**

```python
data = [1, 4, 5, 2, 7, 21,     \
        31, 41, 3, 40, 5, 14,  \
        9, 32, 12, 18, 1, 30,  \
        6, 19, 23, 35, 12, 13, \
        0, 12, 43, 41, 11, 9]

def map_reduce(data, num_processes, mapper, reducer):
    chunks = make_chunks(data, num_processes)
    with Pool(num_processes) as pool:
        chunk_results = pool.map(mapper, chunks)
    return functools.reduce(reducer, chunk_results)

max_value = map_reduce(data, 2, max, max)
```

**Task**

Our goal is to implement the `mapper()` and `reducer()` functions and apply the `map_reduce()` function. This time, we'll break the `job_postings` DataFrame into chunks.



**Answer**

```python
import pandas as pd

job_postings = pd.read_csv("DataEngineer.csv")
job_postings["Job Description"] = job_postings["Job Description"].str.lower()

skills = pd.read_csv("Skills.csv")


def mapper(jobs_chunk):
    frequency = {}
    for skill_name in skills["Name"]:
        frequency[skill_name] = jobs_chunk["Job Description"].str.count(skill_name).sum()
    return frequency

def reducer(freq_chunk1, freq_chunk2):
    merged = {}
    for skill in freq_chunk1:
        merged[skill] = freq_chunk1[skill] + freq_chunk2[skill]
    return merged
    
skill_freq = map_reduce(job_postings, 1, mapper, reducer)
```

In this file, we learned about MapReduce and implemented our own framework. To process data with MapReduce we need to do the following:

1. Define a **mapper** function that takes one chunk of data as input and processes it.
2. Defined a **reducer** function that takes two processes chunks as input and merges the results.

The major advantage of this approach is that, as a user, we don't need to worry at all about parallel processing.

We can't solve every calculation with the MapReduce framework. Some calculations require the entire dataset. MapReduce only works when we can deduce the overall answer from partial chunk results. On the next file, we'll practice solving problems with MapReduce.
